```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Titanic_Preprocessed.csv')
```

Survived is the target variable. Before building a model, we should know whether the classes are balanced or imbalanced.

Count Plot:Because Survived is a categorical variable (0 or 1).

```python
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()
```

Histograms show the distribution of a numerical variable.

```python
sns.histplot(df['Age'], kde=True)
```

```python
sns.boxplot(y=df['Fare'])
```

```python
sns.kdeplot(df['Age'])
```

Survival by Gender

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Gender")
plt.show()
```

0---male,blue:died,orange :alive

1--female

from above observe that females had a higher survival rate than males.

Survival by Passenger Class

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival by Passenger Class")
plt.show()
```

First-class passengers generally survived at a higher rate than third-class passengers.

piechart

```python
df['Sex'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)
plt.show()
```

embarked vs survival:   Different boarding ports may have different passenger characteristics.

```python
sns.countplot(
    x='Embarked',
    hue='Survived',
    data=df
)
plt.show()
```

Passengers boarding at Cherbourg generally showed a higher survival rate than those boarding at Southampton

To see if family companionship influenced survival.

```python
sns.countplot(x='SibSp', hue='Survived', data=df)
plt.show()
```

Passengers traveling with one or two siblings/spouses often had better survival than those traveling alone or in very large groups.

To analyze family relationships and survival.

```python
sns.countplot(x='Parch', hue='Survived', data=df)
plt.show()
```

```python
sns.scatterplot(x='Age', y='Fare', data=df)
```

Multivariate Analaysis

```python
sns.pairplot(df, hue='Survived')
```

```python
sns.violinplot(x='Pclass', y='Age', hue='Survived', data=df)
```
